```python
import numpy as np
import matplotlib.pyplot as plt

ff = 'RC'
## read data
pos = np.genfromtxt("/Users/dian/Documents/2ndProject/Homberger/%s1_6_1.txt" % ff)
number = pos[:,0]
x = pos[:,1]
y = pos[:,2]
dim = len(x)
```

```python
## type 1: original (x,y)
fig, ax = plt.subplots( nrows=1, ncols=1 )
plt.plot(x, y, 'ko', markersize=2)
plt.axis([0, dim/2, 0, dim/2])
plt.grid(True)
plt.show()
fig.savefig('/Users/dian/Documents/2ndProject/Homberger/%s601.png' % ff)
out = open("/Users/dian/Documents/2ndProject/Homberger/%s601.txt" % ff, "w")
for i in range(dim):
    out.write("%d\t%d\t%d\n" % (i, x[i], y[i]))
```

```python
## type 2: reflections to y-axis
fig, ax = plt.subplots( nrows=1, ncols=1 )
xx = -x+dim/2
yy = y
plt.plot(xx, yy, 'ko', markersize=2)
plt.axis([0, dim/2, 0, dim/2])
plt.grid(True)
plt.show()
fig.savefig('/Users/dian/Documents/2ndProject/Homberger/%s602.png' % ff)
out = open("/Users/dian/Documents/2ndProject/Homberger/%s602.txt" % ff, "w")
for i in range(dim):
    out.write("%d\t%d\t%d\n" % (i, xx[i], yy[i]))
```

```python
## type 3: reflections to x-axis
fig, ax = plt.subplots( nrows=1, ncols=1 )
xx = x
yy = -y+dim/2
plt.plot(xx, yy, 'ko', markersize=2)
plt.axis([0, dim/2, 0, dim/2])
plt.grid(True)
plt.show()
fig.savefig('/Users/dian/Documents/2ndProject/Homberger/%s603.png' % ff)
out = open("/Users/dian/Documents/2ndProject/Homberger/%s603.txt" % ff, "w")
for i in range(dim):
    out.write("%d\t%d\t%d\n" % (i, xx[i], yy[i]))
```

```python
## type 4: rotation 180 Clockwise, reflections to O(0,0)
fig, ax = plt.subplots( nrows=1, ncols=1 )
xx = -x+dim/2
yy = -y+dim/2
plt.plot(xx, yy, 'ko', markersize=2)
plt.axis([0, dim/2, 0, dim/2])
plt.grid(True)
plt.show()
fig.savefig('/Users/dian/Documents/2ndProject/Homberger/%s604.png' % ff)
out = open("/Users/dian/Documents/2ndProject/Homberger/%s604.txt" % ff, "w")
for i in range(dim):
    out.write("%d\t%d\t%d\n" % (i, xx[i], yy[i]))
```
